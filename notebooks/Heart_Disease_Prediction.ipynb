```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import ipywidgets as widgets
from IPython.display import display
```

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
```

```python
age = widgets.FloatText(description="Age:")
sex = widgets.Dropdown(
    options=[('Male', 0), ('Female', 1)],
    description='Sex:'
)
cp = widgets.FloatText(description="CP:")
trestbps = widgets.FloatText(description="Trestbps:")
chol = widgets.FloatText(description="Chol:")
fbs = widgets.FloatText(description="FBS:")
restecg = widgets.FloatText(description="Restecg:")
thalach = widgets.FloatText(description="Thalach:")
exang = widgets.Dropdown(
    options=[('No', 0), ('Yes', 1)],
    description='Exang:'
)
oldpeak = widgets.FloatText(description="Oldpeak:")
slope = widgets.FloatText(description="Slope:")
ca = widgets.FloatText(description="CA:")
thal = widgets.FloatText(description="Thal:")

button = widgets.Button(description="Predict")
output = widgets.Output()
```

```python
def on_button_click(b):
    with output:
        input_data = [
            age.value,
            sex.value,
            cp.value,
            trestbps.value,
            chol.value,
            fbs.value,
            restecg.value,
            thalach.value,
            exang.value,
            oldpeak.value,
            slope.value,
            ca.value,
            thal.value  # Add this line for thal
        ]

        # Convert input data to a numpy array and reshape it
        input_data_as_numpy_array = np.asarray(input_data)
        input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

        # Make a prediction
        prediction = model.predict(input_data_reshaped)

        if prediction[0] == 0:
            result = 'The Person does not have a Heart Disease'
        else:
            result = 'The Person has Heart Disease'

        # Clear previous output and display the result
        output.clear_output()
        print(result)

button.on_click(on_button_click)
```

**Heart Disease Prediction Form: Field Descriptions**

1. **age**: Type in the patient's age.

2. **sex**: Choose the patient's sex (Male or Female) from a dropdown list.

3. **cp**: Enter the type of chest pain the patient experiences.

4. **trestbps**: Type in the patient's resting blood pressure (measured in mm Hg).

5. **chol**: Enter the patient's cholesterol level (measured in mg/dl).

6. **fbs**: Indicate whether the patient’s fasting blood sugar level is above 120 mg/dl (No or Yes).

7. **restecg**: Enter the result of the patient's resting electrocardiogram (Normal or Abnormal).

8. **thalach**: Type in the maximum heart rate the patient achieved during exercise.

9. **exang**: Choose whether the patient had exercise-induced angina (No or Yes).

10. **oldpeak**: Enter the level of depression experienced during exercise compared to rest.

11. **slope**: Enter the slope of the ST segment on the heart's exercise test (Upsloping, Flat, or Downsloping).

12. **ca**: Type in the number of major blood vessels that were colored during a fluoroscopy test.
